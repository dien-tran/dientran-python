```python
import matplotlib.pyplot as plt

def CreateALineChart():
    x=[1,2,3,4,5]
    y=[2,4,5,9,12]
    plt.plot(x,y)
    plt.title("Line chart")
    plt.show()

CreateALineChart()
```

```python
def CreateAScatterPlot():
    x=[4,6,2,8,3,12,14]
    y=[2,4,6,8,12,14,90]
    plt.scatter(x,y)
    plt.show()

CreateAScatterPlot()
```

```python
def bieuDoCot():
    category= ['A','B','C','D','E','F']
    values=[3,6,1,2,10,8]
    plt.bar(category,values)
    plt.title("Bar chart")
    plt.show()

bieuDoCot()
```

```python
import numpy as np
def histogram():
    data=np.random.randn(1000)
    plt.hist(data,bins=100)
    plt.title("Histogram")
    plt.show()

histogram()
```

```python
def bieuDoTron():
    size=[10,30,25,15,20]
    lable=['A','B','C','D','E']
    plt.pie(size,labels=lable, autopct='%1.1f%%')
    plt.title("Biu  trn")
    plt.show()
bieuDoTron()
```

```python
def bieuDoChong():
    x=[1,2,3,4,5]
    y1=[2,4,5,9,12]
    y2=[2,4,5,9,12]
    plt.stackplot(x,y1,y2, labels=["y1","y2"])
    plt.legend(["y1","y2"])
    plt.show()

bieuDoChong()
```

```python
import numpy as np
def bieuDoHop():
    # To 3 dataset khc nhau cho boxplot
    data1 = np.random.randn(100)
    data2 = np.random.randn(100)
    data3 = np.random.randn(100)

    data = [data1, data2, data3]

    plt.boxplot(data)
    plt.title("Biu  hp (Box Plot)")
    plt.ylabel("Gi tr")
    plt.xlabel("Dataset")
    plt.show()

bieuDoHop()
```

```python
def bieuDoViolin():
    data1 = np.random.randn(100)
    data2 = np.random.randn(100)
    data3 = np.random.randn(100)

    data = [data1, data2, data3]

    plt.violinplot(data)
    plt.title("Biu  violin")
    plt.ylabel("Gi tr")
    plt.xlabel("Dataset")
    plt.show()

bieuDoViolin()
```

```python
def bieuDoMauNhiet():
    data = np.random.rand(10, 10)
    plt.imshow(data, cmap='hot', interpolation='nearest')
    plt.colorbar()
    plt.title("Biu  mu nhit")
    plt.show()
bieuDoMauNhiet()
```

```python
def bieuDoDuongCong():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.plot(x, y)
    plt.title("Biu  ng cong")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.grid()
    plt.show()
bieuDoDuongCong()
```
