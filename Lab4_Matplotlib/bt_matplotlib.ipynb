```python
import matplotlib.pyplot as plt

def CreateALineChart():
    # Dữ liệu mẫu: chiều cao (cm) và cân nặng (kg)
    heights = [150, 160, 170, 180, 190]
    weights = [50, 60, 65, 75, 85]
    # Vẽ đường với marker 'o' để hiển thị từng điểm
    plt.plot(heights, weights, marker='o') #trục x và trục y với các điểm được thể hiện là chấm o
    plt.title("Line chart")
    plt.xlabel("Chiều cao")
    plt.ylabel("Cân nặng")
    plt.show()

CreateALineChart()
```

```python
def CreateAScatterPlot():
    heights = [150, 160, 170, 180, 190]
    weights = [50, 60, 65, 75, 85]
    plt.scatter(heights, weights, marker='o') #trục x và trục y với các điểm được thể hiện là chấm o
    plt.title("Scatter chart")
    plt.xlabel("Chiều cao")
    plt.ylabel("Cân nặng")
    plt.show()

CreateAScatterPlot()

```

```python
def Barchart():
    heights = [140, 150, 160, 170, 180, 190]
    values=[3,6,1,2,10,8]
    plt.bar(heights,values)
    plt.title("Bar chart")
    plt.xlabel("Chiều cao")
    plt.ylabel("Số lượng")
    plt.show()

Barchart()
```

```python
import numpy as np
def histogram():
    data=np.random.randn(1000)
    plt.hist(data,bins=100) #chia thành 100 cột
    plt.title("Histogram")
    plt.show()

# Biểu đồ tần suất : 
# Mục đích: giúp thấy hình dạng phân phối (ví dụ: chuẩn, lệch phải/trái, nhiều đỉnh), 
# phát hiện outlier, ước lượng mật độ xác suất gần đúng.
histogram()
```

```python
def bieuDoTron():
    size=[10,30,25,15,20] #danh sách kích thước các phần
    lable=['A','B','C','D','E']   #danh sách nhãn các phần
    plt.pie(size,labels=lable, autopct='%1.1f%%')  #vẽ biểu đồ tròn với phần trăm hiển thị trên từng phần
    plt.title("Biểu đồ tròn")
    plt.show()
bieuDoTron()
```

```python
def bieuDoChong():
    x=[1,2,3,4,5]
    y1=[2,4,5,9,12]
    y2=[2,4,5,9,12]
    plt.stackplot(x,y1,y2, labels=["y1","y2"])
    plt.legend(["y1","y2"])
    plt.show()

bieuDoChong()
```

```python
import numpy as np
def bieuDoHop():
    # Tạo 3 dataset khác nhau cho boxplot
    data1 = np.random.randn(100)
    data2 = np.random.randn(100)
    data3 = np.random.randn(100)

    data = [data1, data2, data3]

    plt.boxplot(data)
    plt.title("Biểu đồ hộp (Box Plot)")
    plt.ylabel("Giá trị")
    plt.xlabel("Dataset")
    plt.show()

bieuDoHop()
```

```python
def bieuDoViolin():
    data1 = np.random.randn(100)
    data2 = np.random.randn(100)
    data3 = np.random.randn(100)

    data = [data1, data2, data3]

    plt.violinplot(data)
    plt.title("Biểu đồ violin")
    plt.ylabel("Giá trị")
    plt.xlabel("Dataset")
    plt.show()

bieuDoViolin()
```

```python
def bieuDoMauNhiet():
    data = np.random.rand(10, 10)
    plt.imshow(data, cmap='hot', interpolation='nearest')
    plt.colorbar()
    plt.title("Biểu đồ màu nhiệt")
    plt.show()
bieuDoMauNhiet()
```

```python
def bieuDoDuongCong():
    x = np.linspace(0, 10, 100)
    y = np.sin(x)
    plt.plot(x, y)
    plt.title("Biểu đồ đường cong")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.grid()
    plt.show()
bieuDoDuongCong()
```
