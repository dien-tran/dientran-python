```python
import numpy as np
```

```python
a = np.random.randint(0,10,(3,3)) 
b = np.random.randint(0,10,(3,3)) 
c = np.random.randint(0,10,(2,2))
```

```python
def TongMatran(a,b):
    if a.shape != b.shape: # nếu 2 ma trận ko cùng kích thước
        return "Không cộng được"
    print (a + b)

TongMatran(a,b)
```

```python
def HieuMatran(a,b):
    if a.shape != b.shape: # nếu 2 ma trận ko cùng kích thước
        return "Không trừ được"
    return (a - b)

HieuMatran(np.array(a),np.array(b))
```

```python
def NhanMatran(a,b):
    if a.shape != b.shape: # nếu 2 ma trận ko cùng kích thước
        return "Không nhân được"
    return (a * b)

NhanMatran(np.array(a),np.array(b))
```

```python
def NhanMaTranVoiSoTuNhien(a,n):
    return a * n

print(a)
NhanMaTranVoiSoTuNhien(np.array(a),3)
```

```python
def ChiaMatran(a,b):
    if a.shape != b.shape: # nếu 2 ma trận ko cùng kích thước
        return "Không chia được"
    return (a / b)

ChiaMatran(np.array(a),np.array(b))
```

```python
def ChiaMaTranVoiSoTuNhien(a,n):
    return a / n

print(a)
ChiaMaTranVoiSoTuNhien(np.array(a),3)
```

```python
def NoiMaTran(a,b):
    if a.shape[0] != b.shape[0]: # nếu 2 ma trận ko cùng số hàng
        return "Không nối được"
    return np.concatenate((a,b), axis=1) # Nối theo chiều ngang

print(a)
print(b)
NoiMaTran(np.array(a),np.array(b))
```

```python
def NoiMaTran(a,b):
    if a.shape[0] != b.shape[0]: # nếu 2 ma trận ko cùng số hàng
        return "Không nối được"
    return np.concatenate((a,b), axis=0) # Nối theo chiều doc

print(a)
print(b)
NoiMaTran(np.array(a),np.array(b))
```

```python
print(a)
print('\n')

print("in hàng 1 và 2 + in cột 1 và 2")
print(a[0:2,0:3]) # 0:2 là in hàng 1 và 2 + 0:3 là in cột 1 và 2
print('\n')

print("in tất cả các hàng + in cột 1 và 2")
print(a[:,0:2]) # : là in tất cả các hàng + 1:3 là in cột 1 và 2
```

```python
# tính định thức bậc 2 và bậc 3 

def DetBac2(c):
    return c[0,0]*c[1,1] - c[0,1]*c[1,0]
    
def DetBac2Ham(c):
    return np.linalg.det(c)

print(c)
print(DetBac2(np.array(c)))
print(DetBac2Ham(np.array(c)))

        
```
